1. (6б). Створіть клас Building, додайте три атрибути, один з яких має значення за замовчуванням та два методи на свій розсуд.

```python
class Building:
  def __init__(self, building_type, year_built, num_of_storeys = 5):
    self._building_type = building_type
    self._year_built = year_built
    self._num_of_storeys = num_of_storeys

  @property
  def building_type(self):
    return self._building_type

  @property
  def year_built(self):
    return self._year_built

  @property
  def num_of_storeys(self):
    return self._num_of_storeys

  def build_more_floors(self, storeys_to_add = 1):
    if self._year_built < 2000:
      print("Sorry, the building is too old for that")
      return

    print(f"Building {storeys_to_add} more storeys...")
    self._num_of_storeys += storeys_to_add

  def destroy_building(self):
    print(f"Destroying the building {self}")
    self._building_type = "Ruins"
    self._num_of_storeys = 0
    print("\nBuilding destroyed successfully")

  def __str__(self):
    return f"{self._building_type} from {self._year_built} ({self._num_of_storeys} storeys)"
```

5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.



```python
class ThreeDimFigure:
  def get_volume(self):
    return 0

  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

```python
class Cylinder(ThreeDimFigure):
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def get_volume(self):
    return 3.14 * (self.radius ** 2) * self.height
```

```python
class Cube(ThreeDimFigure):
  def __init__(self, length):
    self.length = length

  def get_volume(self):
    return self.length ** 3
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
    def __init__(self, id, balance=0):
        self.__id = id
        self.__balance = balance

    @property
    def id(self):
        return self.__id

    @property
    def balance(self):
        return self.__balance

    @balance.setter
    def balance(self, value):
        if value < 0:
            raise ValueError("Balance cannot be negative")
        self.__balance = value

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be positive")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be positive")
        if amount > self.__balance:
            raise ValueError("Insufficient funds")
        self.__balance -= amount
```

```python
wallet = EWallet(1, 2000000)
```

```python
wallet.balance
```

```python
wallet.balance = 100000
```

```python
wallet.deposit(10000)
wallet.balance
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class ShopItem:
  def __init__(self, name, price):
    self.name = name
    self.price = price

  def __str__(self):
    return f"{self.name} (${self.price})"
```

```python
class ShoppingCart:
  def __init__(self, items: list[ShopItem]):
    self.items = items

  def add_item(self, item: ShopItem):
    self.items.append(item)

  def remove_item(self, item: ShopItem):
    self.items.remove(item)

  def show_items(self):
    for item in self.items:
      print(item)

  def total_price(self):
    total_price = 0
    for item in self.items:
      total_price += item.price

    return total_price
```

```python
item_milk = ShopItem("Milk", 5)
item_bread = ShopItem("Bread", 2.5)
item_meat = ShopItem("Chicken Breasts", 40)
```

```python
shopping_cart = ShoppingCart([item_milk, item_bread, item_meat])
```

```python
shopping_cart.add_item(item_bread)
```

```python
shopping_cart.show_items()
```

```python
shopping_cart.remove_item(item_meat)
```

```python
shopping_cart.total_price()
```

Ітерабельні обʼєкти - це обʼєкти, по яких можна проітеруватися за допомогою ітератора. В собі вони мають метод \_\_iter__, таким чином повертаючи потрібен ітератор, коли по них треба проітеруватися.

Ітератор - це екземпляр класу, який має методи \_\_iter__ та \_\_next__, відповідно знає, який елемент повертати наступним,. що дозволяє йому керувати процесом ітерування.

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekIter:
    def __init__(self):
        self.days = [
            "Monday", "Tuesday", "Wednesday",
            "Thursday", "Friday", "Saturday", "Sunday"
        ]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.days):
            raise StopIteration
        day = self.days[self.index]
        self.index += 1
        return day
```

```python
week = WeekIter()
```

```python
next(week)
```

```python
next(week)
```
